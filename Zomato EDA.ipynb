# Zomato Exploratory Data Analysis

```python
# Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Read the dataset
df = pd.read_csv("zomato.csv")
df.head()
```

```python
#Total no. of rows & colums
df.shape
```

```python
#All the column names
df.columns
```

```python
#Removing unwanted columns
df=df.drop(['url','address','phone','dish_liked','reviews_list','menu_item'],axis=1)
df.head()
```

```python
df.info()
```

```python
#Remove duplicate columns
df.drop_duplicates(inplace=True)
df.shape
```

```python
#All unique values in 'Rate' column
df['rate'].unique()
```

```python
#Clean 'Rate' column
def cleanrate(value):
    if(value=='NEW' or value =='-'):
        return np.nan
    else:
        value = str(value).split('/')
        value = value[0]
        return float(value)
    
df['rate']= df['rate'].apply(cleanrate)
df['rate'].head()
```

```python
#Total null values in 'Rate' column
df['rate'].isnull().sum()
```

```python
#Replace null values with Mean
df['rate'].fillna(df['rate'].mean(),inplace=True)
df['rate'].isnull().sum()
```

```python
#All information about dataset
df.info()
```

```python
#Remove all null values
df.dropna(inplace=True)
df.head()
```

```python
#Rename columns
df.rename(columns ={'approx_cost(for two people)':'Cost4two','listed_in(type)':'Type'},inplace=True)
df.head()
```

```python
#Unique values in 'Location; column
df['location'].unique()
```

```python
#Unique values in 'Listed_in(city)' column
df['listed_in(city)'].unique()
```

```python
#Drop 'Listed_in(city)' column
df=df.drop(['listed_in(city)'],axis=1)
```

```python
df.head()
```

```python
#Unique values in 'Cost4two' values
df['Cost4two'].unique()
```

```python
#Remove comma from 'Cost4two' values and change datatype
def removecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['Cost4two'] = df['Cost4two'].apply(removecomma)
df['Cost4two'].unique()
```

```python
df.head()
```

```python
#No. of different restaurant types
df['rest_type'].value_counts()
```

```python
rest_type=df['rest_type'].value_counts()
rest_type_lessthan1000 = rest_type[rest_type<1000]
```

```python
def manage_rest_type(value):
    if(value in rest_type_lessthan1000):
        return'Others'
    else:
        return value
    
df['rest_type'] = df['rest_type'].apply(manage_rest_type)
df['rest_type'].value_counts()
```

```python
df['location'].value_counts()
```

```python
location=df['location'].value_counts()
location_lessthan250 = location[location<250]
```

```python
def manage_location(value):
    if(value in location_lessthan250):
        return'Others'
    else:
        return value
    
df['location'] = df['location'].apply(manage_location)
df['location'].value_counts()
```

```python
df.head()
```

```python
cuisine=df['cuisines'].value_counts()
cuisines_lessthan100 = cuisine[cuisine<100]
```

```python
def manage_cuisine(value):
    if(value in cuisines_lessthan100):
        return 'Others'
    else:
        return value
    
df['cuisines'] = df['cuisines'].apply(manage_cuisine)
df['cuisines'].value_counts()
```

```python
df.head()
```

```python
df['Type'].value_counts()
```

# Data Visualization

```python
#Restaurants delivering Online or Not(Countplot)
plt.figure(figsize=(6,6))
sns.countplot(df['online_order'],palette = 'inferno')
plt.title ("Restaurants delivering Online or Not")
```

```python
#Restaurants allowing Table booking or Not(Countplot)
plt.figure(figsize=(6,6))
sns.countplot(df['book_table'],palette = 'rainbow')
plt.title ("Restaurants allowing Table booking or Not")
```

```python
#Types of Restaurants(Countplot)
plt.figure(figsize=(6,6))
sns.countplot(df['rest_type'],palette = 'magma')
plt.xticks(rotation = 90)
plt.title ("Types of Restaurants")
```

```python
#Restaurants in different locations(Countplot)
plt.figure(figsize=(10,10))
sns.countplot(df['location'])
plt.xticks(rotation = 90)
plt.title ("Restaurants in different locations")
```

```python
#Online Order Facility vs Ratings(Boxplot)
plt.figure(figsize=(6,6))
sns.boxplot(x='online_order',y='rate',data=df,palette='rainbow')
plt.title ("Online Order Facility vs Ratings")
```

```python
df1=df.groupby(['location','online_order'])['name'].count()
df1.to_csv('location_online.csv')
df1=pd.read_csv('location_online.csv')
df1=pd.pivot_table(df1,values=None,index=['location'],columns=['online_order'],fill_value=0,aggfunc=np.sum)
df1
```

```python
df1.plot(kind='bar', figsize=(16,8))
```

```python
#Book Table Facility vs Ratings(Boxplot)
plt.figure(figsize=(6,6))
sns.boxplot(x='book_table',y='rate',data=df,palette='magma')
plt.title ("Book Table Facility vs Ratings")
```
